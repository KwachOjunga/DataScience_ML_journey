```python
#Reading Data into DFs
train_data = pd.read_csv("Data/Datasets/Multi_class_obesity_prediction/train.csv(1)/train.csv")
test_data = pd.read_csv("Data/Datasets/Multi_class_obesity_prediction/test.csv(1)/test.csv")

#Viewing Data
train_data.head()
```

```python
#Cleaning data and altering to apppropriate formats
train_data["Gender_male"] = train_data["Gender"] == 'Male'
train_data["family_history_with_overweight_true"] = train_data["family_history_with_overweight"] == 'yes'
train_data["FAVC_yes"] = train_data["FAVC"] == 'yes'
train_data["Smoker"] = train_data["SMOKE"] == 'yes'
train_data["SCC_yes"] = train_data["SCC"] =='yes'

test_data["Gender_male"] = test_data["Gender"] == 'Male'
test_data["family_history_with_overweight_true"] = test_data["family_history_with_overweight"] == 'yes'
test_data["FAVC_yes"] = test_data["FAVC"] == 'yes'
test_data["Smoker"] = test_data["SMOKE"] == 'yes'
test_data["SCC_yes"] = test_data["SCC"] =='yes'

train_data = train_data.drop("Gender",axis=1)
train_data = train_data.drop("family_history_with_overweight",axis=1)
train_data = train_data.drop("FAVC",axis=1)
train_data = train_data.drop("SMOKE",axis=1)
train_data = train_data.drop("SCC",axis=1)

test_data = test_data.drop("Gender",axis=1)
test_data = test_data.drop("family_history_with_overweight",axis=1)
test_data = test_data.drop("FAVC",axis=1)
test_data = test_data.drop("SMOKE",axis=1)
test_data = test_data.drop("SCC",axis=1)

test_data = pd.get_dummies(test_data,columns=["MTRANS","CALC","CAEC"])
train_data = pd.get_dummies(train_data,columns=["MTRANS","CALC","CAEC"])
```
